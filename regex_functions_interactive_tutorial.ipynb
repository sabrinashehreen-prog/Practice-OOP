```python

import re
from pprint import pprint

def inspect_match(match):
    """Pretty-print a match object."""
    if match is None:
        print("No match")
        return
    print("Matched text :", match.group())
    print("Start index  :", match.start())
    print("End index    :", match.end())
    print("Span         :", match.span())
    print("Groups       :", match.groups())
    print("Group dict   :", match.groupdict())

def show_finditer(pattern, text, flags=0):
    print(f"PATTERN: {pattern!r}")
    print(f"TEXT   : {text!r}")
    matches = list(re.finditer(pattern, text, flags))
    print(f"TOTAL MATCHES: {len(matches)}")
    for i, m in enumerate(matches, start=1):
        print(f"  {i}. {m.group()!r} at span {m.span()}")

def divider(title=""):
    print("\n" + "=" * 80)
    print(title)
    print("=" * 80)
```

## 1) What is a Regular Expression?

A **regular expression (RegEx)** is a pattern used to search, validate, extract, split, or replace text.

Example ideas:
- Find all numbers in a sentence
- Extract emails from text
- Validate whether a string is in a required format
- Replace repeated spaces with one space

```python

text = "My 2 favorite numbers are 19 and 42."
pattern = r"\d+"

print("Text   :", text)
print("Pattern:", pattern)
print("Matches:", re.findall(pattern, text))
```

## 2) Quick RegEx symbols

Common symbols:
- `.` → any character except newline
- `\d` → digit
- `\D` → not a digit
- `\w` → word character (`a-z`, `A-Z`, `0-9`, `_`)
- `\W` → not a word character
- `\s` → whitespace
- `\S` → non-whitespace
- `^` → start of string/line
- `$` → end of string/line
- `*` → 0 or more
- `+` → 1 or more
- `?` → 0 or 1
- `{m,n}` → between `m` and `n` repetitions
- `[abc]` → one character from the set
- `[^abc]` → one character not in the set
- `(...)` → capture group
- `|` → OR

```python

samples = [
    (r".", "A"),
    (r"\d+", "Room 407"),
    (r"\w+", "Dr_C"),
    (r"\s+", "a   b"),
    (r"[aeiou]", "education"),
    (r"[^aeiou\s]", "data science"),
]

for pattern, text in samples:
    divider(f"Pattern {pattern!r} on {text!r}")
    print(re.findall(pattern, text))
```

## 3) `re.search()`

`re.search()` looks for the **first place** where the pattern appears anywhere in the string.

```python

text = "Contact us at support@example.com for help."
match = re.search(r"\S+@\S+", text)
inspect_match(match)
```

```python

text = "Order ID: A-493, Status: shipped"
inspect_match(re.search(r"Status:\s+\w+", text))
```

## 4) `re.match()`

`re.match()` checks only from the **beginning** of the string.

```python

text = "Python is powerful"

print("search:", re.search(r"is", text))
print("match :", re.match(r"is", text))
print("match :", re.match(r"Python", text))
```

## 5) `re.fullmatch()`

`re.fullmatch()` succeeds only if the **entire string** matches the pattern.

```python

tests = ["CS3600", "CS-3600", "MATH101", "CS99999"]
pattern = r"[A-Z]{2,4}\d{3,4}"

for t in tests:
    ok = re.fullmatch(pattern, t)
    print(f"{t:10} ->", "VALID" if ok else "INVALID")
```

## 6) `re.findall()`

`re.findall()` returns a **list of all matches**.

```python

text = "Temperatures this week: 72F, 68F, 75F, 70F"
print(re.findall(r"\d+", text))
print(re.findall(r"\d+F", text))
```

```python

text = "Emails: a@x.com, bob@test.edu, admin@site.org"
print(re.findall(r"\b[\w.-]+@[\w.-]+\.\w+\b", text))
```

## 7) `re.finditer()`

`re.finditer()` returns an **iterator of match objects**.

```python

text = "IDs: X12, Y305, Z9"
show_finditer(r"[A-Z]\d+", text)
```

## 8) `re.split()`

```python

text = "apple,banana;orange | mango   grape"
parts = re.split(r"[,;|\s]+", text)
print(parts)
```

## 9) `re.sub()`

```python

text = "My phone numbers are 435-555-1111 and 435-555-2222"
masked = re.sub(r"\d", "X", text)
print(masked)
```

```python

text = "Too     many       spaces"
clean = re.sub(r"\s+", " ", text)
print(clean)
```

## 10) `re.subn()`

```python

text = "cat bat rat mat"
new_text, count = re.subn(r"\b\wat\b", "pet", text)
print("Result      :", new_text)
print("Replacements:", count)
```

## 11) `re.compile()`

```python

email_pattern = re.compile(r"\b[\w.-]+@[\w.-]+\.\w+\b")

texts = [
    "Email alice@school.edu today.",
    "No email here.",
    "Team: bob@org.com and carol@dept.net"
]

for t in texts:
    print(email_pattern.findall(t))
```

## 12) `re.escape()`

```python

user_text = "price is $10.00 (limited-time)"
escaped = re.escape(user_text)

print("Original:", user_text)
print("Escaped :", escaped)
print("Literal match works?:", bool(re.search(escaped, "Today the price is $10.00 (limited-time)!")))
```

## 13) Character classes

### Predefined classes
- `\d`, `\D`
- `\w`, `\W`
- `\s`, `\S`

### Custom classes
- `[abc]`
- `[a-z]`
- `[A-Z0-9]`
- `[^0-9]`

```python

text = "Room B12 is on floor 3."

patterns = [
    r"\d",
    r"\D",
    r"\w+",
    r"\W+",
    r"[A-Z]",
    r"[A-Z]\d+",
    r"[^a-zA-Z0-9\s]+",
]

for p in patterns:
    divider(p)
    print(re.findall(p, text))
```

## 14) Quantifiers

- `*` → zero or more
- `+` → one or more
- `?` → zero or one
- `{m}` → exactly `m`
- `{m,}` → at least `m`
- `{m,n}` → between `m` and `n`

```python

text = "ha haa haaa haaaa"

patterns = [
    r"ha*",
    r"ha+",
    r"ha?",
    r"ha{2}",
    r"ha{2,3}",
]

for p in patterns:
    divider(p)
    print(re.findall(p, text))
```

## 15) Greedy vs non-greedy matching

```python

text = "<title>Intro</title><title>Advanced</title>"

print("Greedy     :", re.findall(r"<title>.*</title>", text))
print("Non-greedy :", re.findall(r"<title>.*?</title>", text))
```

## 16) Anchors: `^` and `$`

```python

lines = [
    "From: alice@example.com",
    "Date: today",
    "From: bob@test.org",
    "Subject: hello"
]

for line in lines:
    if re.search(r"^From:", line):
        print("Matched ->", line)
```

```python

tests = ["abc123", "123abc", "abc123xyz"]
for t in tests:
    print(t, "->", bool(re.search(r"^abc123$", t)))
```

## 17) Groups

```python

text = "From stephen.marquard@uct.ac.za Sat Jan 5"
match = re.search(r"@(\S+)", text)
inspect_match(match)
```

```python

text = "Name: Alice, Age: 24"
match = re.search(r"Name:\s+(\w+),\s+Age:\s+(\d+)", text)

print("Whole match:", match.group(0))
print("Group 1    :", match.group(1))
print("Group 2    :", match.group(2))
print("All groups :", match.groups())
```

### Non-capturing groups
Use `(?:...)` when you want grouping without storing the result as a capture group.

```python

text = "color colour color colour"
print(re.findall(r"colou?r", text))
print(re.findall(r"colo(?:u)?r", text))
```

### Named groups

```python

text = "2026-03-16"
m = re.search(r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})", text)
inspect_match(m)

print("Year :", m.group("year"))
print("Month:", m.group("month"))
print("Day  :", m.group("day"))
```

## 18) Alternation with `|`

```python

text = "The colors are red, blue, green, and black."
print(re.findall(r"red|blue|green", text))
```

## 19) Flags

Useful flags:
- `re.I` or `re.IGNORECASE`
- `re.M` or `re.MULTILINE`
- `re.S` or `re.DOTALL`
- `re.X` or `re.VERBOSE`

```python

text = "Python python PYTHON"

print(re.findall(r"python", text))
print(re.findall(r"python", text, flags=re.IGNORECASE))
```

```python

text = "Start
Middle
End"

print("Without MULTILINE:", re.findall(r"^\w+", text))
print("With MULTILINE   :", re.findall(r"^\w+", text, flags=re.MULTILINE))
```

```python

text = "begin
content
end"

print("Without DOTALL:", re.findall(r"begin.*end", text))
print("With DOTALL   :", re.findall(r"begin.*end", text, flags=re.DOTALL))
```

```python

pattern = re.compile(r'''
    (?P<area>\d{3})   # area code
    [-\s]?            # optional separator
    (?P<prefix>\d{3}) # prefix
    [-\s]?            # optional separator
    (?P<line>\d{4})   # line number
''', re.VERBOSE)

m = pattern.search("Call me at 435-555-0199.")
inspect_match(m)
```

## 20) Useful Match object methods

## 22) Interesting Example — Parse log entries

```python

logs = '''
2026-03-16 09:00:12 INFO Server started
2026-03-16 09:01:55 WARNING High memory usage
2026-03-16 09:03:10 ERROR Database timeout
'''

pattern = re.compile(
    r"(?P<date>\d{4}-\d{2}-\d{2})\s+"
    r"(?P<time>\d{2}:\d{2}:\d{2})\s+"
    r"(?P<level>INFO|WARNING|ERROR)\s+"
    r"(?P<message>.+)"
)

for m in pattern.finditer(logs):
    print(m.groupdict())
```

## 23) Interesting Example — Clean messy text

```python

messy = "This   is   a    sentence ,   with   bad spacing  !"
step1 = re.sub(r"\s+", " ", messy)
step2 = re.sub(r"\s+([,.!?;:])", r"\1", step1)

print("Original:", messy)
print("Cleaned :", step2)
```

## 24) Interesting Example — Validate simple phone formats

```python

phone_pattern = re.compile(r"^(\(?\d{3}\)?[-\s]?)\d{3}[-\s]?\d{4}$")

tests = [
    "4355551234",
    "435-555-1234",
    "(435) 555-1234",
    "435 555 1234",
    "43-555-1234"
]

for t in tests:
    print(f"{t:18} ->", bool(phone_pattern.fullmatch(t)))
```

## 26) Common mistakes students make

1. Forgetting raw strings
2. Using `match()` when they need `search()`
3. Writing greedy `.*` when a non-greedy `.*?` is needed
4. Forgetting to escape special characters
5. Making patterns too broad

```python

text = "file.txt and filext"

print("Wrong :", re.findall(r"file.txt", text))
print("Right :", re.findall(r"file\.txt", text))
```
